## 6. Cost Awareness

### API Pricing (as of 2025)

| Model | Input (per 1M tokens) | Output (per 1M tokens) |
|-------|----------------------|------------------------|
| gpt-4o-mini | $0.15 | $0.60 |
| gpt-4o | $2.50 | $10.00 |
| text-embedding-3-small | $0.02 | - |

### Expected Course Costs

- **Entire course:** ~$5-15 using gpt-4o-mini
- **Per week:** ~$0.30-1.00

### Tips to Minimize Costs

1. Use `gpt-4o-mini` (default in all labs)
2. Keep `max_tokens` low when experimenting
3. Cache responses when possible
4. Set a spending limit in OpenAI dashboard

````python
# Helper function to estimate costs (use in any lab)
def estimate_cost(response, model="gpt-4o-mini"):
    """Estimate cost of an API call."""
    prices = {
        "gpt-4o-mini": {"input": 0.15, "output": 0.60},
        "gpt-4o": {"input": 2.50, "output": 10.00},
    }

    if model not in prices:
        return "Unknown model"

    input_cost = (response.usage.prompt_tokens / 1_000_000) * prices[model]["input"]
    output_cost = (response.usage.completion_tokens / 1_000_000) * prices[model]["output"]

    return {
        "input_tokens": response.usage.prompt_tokens,
        "output_tokens": response.usage.completion_tokens,
        "total_cost_usd": round(input_cost + output_cost, 6)
    }

print("Cost estimation function ready!")
print("Usage: estimate_cost(response)")
````
